##1ο Μέρος_Θεωρητικό

***Περίληψη***


> Στο paper αυτό περιγράφεται μια τεχνική για κωδικοποίηση εικόνας όπου τελεστές πολλαπλών κλιμάκων αλλά πανομοιότυπου σχήματος εξυπηρετούν ως οι βασικές συναρτήσεις. Ουσιαστικά προτείνεται ένας νεός τρόπος "αποσυσχετισμού" των pixel της εικόνας ο οποίος είναι μη-αιτιατός, με απλούς ωστόσο υπολογισμούς.


***Απλή Κατανόηση***


> Αρχικά, εξηγείται μια κατανόηση της έννοιας κωδικοποίησης. Η διαδικασία της κωδικοποίησης μιας εικόνας γίνεται στη διαφορά της πραγματικής τιμής ενός pixel με τη προβλεφθείσα τιμή του, απο προηγουμένως κωδικοποιημένα pixel. Στη κωδικοποίηση με Laplacian Pyramid, η προβλεπόμενη τιμή για κάθε pixel υπολογίζεται ως τοπικό weighted μέσο χρησιμοποιώντας μια ομοιόμορφη τύπου-Γκαουσιανή συνάρτηση κεντραρισμένη στο pixel. Επομένως, το **πρώτο βήμα**, είναι η συνέλιξη της αρχικής μας εικόνας με μια gaussian-τύπου συνάρτηση για να αποκτήσουμε τη g1. Παρόμοια για τη g2,g3,...,gN. Η αλληλουχία των εικόνων αυτών ονομάζεται **Γκαουσιανή πυραμίδα**. Η συνέλιξη γίνεται με ένα kernel 5x5, του οποίου μια γραμμή τιμών μήκους 5 είναι **κανονικοποιημένη**, **συμμετρική** και **εξίσου σημαντική** για τη παραγωγή του επόμενου επιπέδου. Αν ορίσουμε ως **α** της μεσαία τιμή του, αυτή η τιμή παίζει ρόλο στο **σχήμα** που θα έχει η ισοβαρής συνάρτησή μας. Για παράδειγμα, για α = 0.4 έχουμε μια ισοβαρή συνάρτηση που προσεγγίζει αρκετά τη Γκαουσιανή. Αποκτούμε μια αναπαράσταση της συμπίεσης όταν κωδικοποιούμε την διαφορά της επεκτεταμένη εικόνα g1 απο την αρχική g0. Η διαφορά αυτή ονομάζεται "error image" και μια αλληλουχία αυτών δημιουργεί τη **Laplacian Pyramid**. Η αρχική εικόνα μπορεί να ανακτηθεί ακριβώς, επεκτείνοντας και αθροίζοντας όλα τα επίπεδα της Λαπλασιανής πυραμίδας. Η τεχνική της αφαίρεσης μιας προβλεπόμενης τιμής απο κάθε pixel αφαιρεί το μεγαλύτερο μέρος της συσχέτισης ανάμεσα στα pixel. (εξού και ο στόχος να προβλέψουμε προηγούμενα pixel). Αυτή η αποσυσχέτιση έχει ως αποτέλεσμα μειωμένη **διακύμανση** και **εντροπία της εικόνας**. Το πόσο αυτά τα μεγέθη θα μειωθούν εξαρτάται απο τη τιμή της παραμέτρου α. Γενικά τα επίπεδα της Γκαουσιανής πυραμίδας φάινονται πιο "ευκρινή" όταν παράγονται με αυτή τη τιμή του α.



-***Πώς η τιμή του α επηρεάζει την γκαουσιανή πυραμίδα;***



> Η τιμή αυτή παίζει, αρχικά, ρόλο στο **σχήμα** που θα έχει η ισοβαρής συνάρτησή μας. Αναλόγως το α η συνάρτηση παίρνει διαφορετικό σχήμα. Αυτό επηρεάζει την ευκρίνεια που θα έχουν τα επίπεδα στη Γκαουσιανή πυραμίδα.Επηρεάζει την ευκρίνεια των εικόνων, διότι το α που θα επιλέξουμε, καθορίζει το πόσο θα μειωθούν η διακύμανση και η εντροπία στην εικόνα.

-***Τι ορίζουμε ως εντροπία; Υπολογίστε τη μέγιστη τιμή εντροπίας που μπορεί να έχει μια grayscale εικόνα***


> Η εντροπία είναι ένα μέτρο του αριθμού των bits per pixel που απαιτείται για την ακριβή κωδικοποίηση της εικόνας. Μεγάλη τιμή εντροπίας έχει ως αποτέλεσμα πιο λεπτομερή εικόνα. Αν ορίσουμε ως f(i) τη συχνότητα εμφάνισης του επιπέδου i φωτεινότητας στην εικόνα τότε η εντροπία ορίζεται ως: H = –[Σ (f(i) log2 f(i))] για i απο 0 έως 255. (grayscale values). Η μέγιστη εντροπίας μιας εικόνας εφόσον έχουμε 256 τιμές επιπέδων φωτεινότητας υπολογίζεται ως 8.

-***Πώς η επιλογή του μεγέθους του bin επηρεάζει το αποτέλεσμα του κβαντισμού;***


> Η εντροπία μπορεί να μειωθεί σημαντικά με τη κβαντοποίηση των τιμών pixel σε κάθε επίπεδο της Λαπλασιανής πυραμίδας. Το μέγεθος του bin σε ένα δοσμένο επίπεδο αντικατοπτρίζει την ευαισθησία του ανθρώπινου παρατηρητή να αναγνωρίζει λάθη εντός της χωρικής συχνότητας που υπάρχει στο επίπεδο. Περισσότερη κβαντοποίηση πρέπει να χρησιμοποιηθεί σε υψηλότερα επίπεδα πυραμίδας απότι σε μικρότερα.

-***Εξηγήστε πώς επηρεάζεται η κβάντιση από το πλήθος των επιπέδων της πυραμίδας;***


>Οι άνθρωποι είναι ευαίσθητοι σε αλλαγές στην αντίθεση της εικόνας σε χαμηλές και μεσαίες συχνότητες δειγμάτων(υψηλά επίπεδα πυραμίδας), αλλά όχι πολύ ευαίσθητοι σε υψηλές χωρικές συχνότητες (χαμηλά επίπεδα πυραμίδας). Αυτό σημαίνει ότι στα υψηλά επίπεδα πρέπει να χρησιμοποιηθούν πολλά επίπεδα κβάντισης (μεγάλη λεπτομέρεια) απότι στα χαμηλά επίπεδα όπου επειδή είναι πολυ πυκνά (υψηλή συχνότητα), μπορούν να κβαντοποιηθούν πιο πρόχειρα και όχι με τόσο λεπτομέρεια. Επομένως όσο πιο πολλά επίπεδα πυραμίδας έχουμε τόσο πιο πολύ πρέπει να κβαντοποιηθεί η εικόνα στα επίπεδα αυτά λόγω χαμηλής συχνότητας δειγμάτων.

```python
#Εισάγουμε την εικόνα lena

HTTP_response = urllib.request.urlopen('http://www.image.ntua.gr/~tpar/LABimage/lena.png')

arr = np.asarray(bytearray(HTTP_response.read()), dtype=np.uint8)

BGR = cv2.imdecode(arr,-1)

lena_img = cv2.cvtColor(BGR, cv2.COLOR_BGR2YCrCb)[:,:,0]
#tmp = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
#Y = tmp

camera_img = data.camera()

fig=plt.figure(dpi=150)

fig.add_subplot(1,2,1)
plt.imshow(lena_img,cmap='gray')
plt.axis("off")
plt.title("Lena")

fig.add_subplot(1,2,2)
plt.imshow(camera_img,cmap='gray')
plt.axis("off")
plt.title("Camera")
```

Παρακάτω περιέχονται όλες οι απαιτούμενες συναρτήσεις για τον αλγόριθμο κωδικοποίησης.

```python
def GKernel(a):
  #Creates kernel h based on a
  b = 1/4
  c = 1/4 - a/2
  kernel1D = [c,b,a,b,c]
  h = np.array(kernel1D)

  return h 


def GREDUCE(I, h):
  #Convolving image I with kernel h and downsampling it

  g = cv2.filter2D(I, -1, h, h)
  I_out = cv2.resize(g, (g.shape[1]//2, g.shape[0]//2), interpolation=cv2.INTER_AREA)

  return I_out


def GPyramid(I, a, depth):

  #Takes image I, kernel with parameter a and a depth for the pyramid
  #returns the gaussian pyramid G

  kernel = GKernel(a)
  G = [I]
  for n in range(depth):
    g = GREDUCE(G[n], kernel)
    G.append(g)

  return G


def GEXPAND(I, h):


  g = cv2.resize(I, (I.shape[1]*2, I.shape[0]*2), interpolation=cv2.INTER_AREA)
  I_out = cv2.filter2D(g, -1, h, h)
  

  return I_out


def LPyramid(I, a, depth):

  #LPyramid will return Laplacian Pyramid of image I

  G = GPyramid(I, a, depth) #GPyramid of image I
  kernel = GKernel(a) 

  L = [G[depth-1]] 
  for n in range(depth-1):
    interpolation = GEXPAND(G[depth-n-1], kernel)
    difference = G[depth-n-2] - interpolation + 128
    L.append(difference)
  L.reverse()

  return L


def L_Pyramid_Decode(L,a):

  #Using Laplacian Pyramid of an image and parameter a
  #used for creating it, the output is the decoded image
  N = len(L) #Depth of the Pyramid
  kernel = GKernel(a) #kernel

  I_out = L[N-1]
  for n in range(N-1):
    I_out = GEXPAND(I_out, kernel)
    I_out = I_out + L[N-n-2] - 128


  return I_out


def L_Quantization(L, n):

  #Based on size of bins n, this function return the quantized value
  #for every pixel of a given level of the pyramid

  C = np.zeros(L.shape)
  
  #C list are the quantized levels

  for x in range(L.shape[0]):
    for y in range(L.shape[1]):
      m = math.ceil(L[x][y]/n - 1/2)
      C[x][y] = m*n
  
  return C
```

**3.Δημιουργία Γκαουσιανής Πυραμίδας**

```python
def GPyramid(I, a, depth):

  #Takes image I, kernel with parameter a and a depth for the pyramid
  #returns the gaussian pyramid G

  kernel = GKernel(a)
  G = [I]
  for n in range(depth-1):
    g = GREDUCE(G[n], kernel)
    G.append(g)

  return G

depth = 4
G = GPyramid(lena_img, a, depth)

fig=plt.figure(dpi=180)
fig.subplots_adjust(hspace=0.4, top=0.9)
fig.suptitle('Gaussian Pyramid')

fig.add_subplot(2,4,1)
plt.imshow(G[0],cmap='gray')
plt.axis("off")
plt.title("Level 0")

fig.add_subplot(2,4,2)
plt.imshow(G[1],cmap='gray')
plt.axis("off")
plt.title("Level 1")

fig.add_subplot(2,4,3)
plt.imshow(G[2],cmap='gray')
plt.axis("off")
plt.title("Level 2")

fig.add_subplot(2,4,4)
plt.imshow(G[3],cmap='gray')
plt.axis("off")
plt.title("Level 3")
```

**4. Δημιουργία Λαπλασιανής πυραμίδας με χρήση του EXPAND**

```python
def LPyramid(I, a, depth):

  #LPyramid will return Laplacian Pyramid of image I

  G = GPyramid(I, a, depth) #GPyramid of image I
  kernel = GKernel(a) 

  L = [G[depth-1]] 
  for n in range(depth-1):
    interpolation = GEXPAND(G[depth-n-1], kernel)
    difference = G[depth-n-2] - interpolation + 128
    L.append(difference)
  L.reverse()

  return L

Laplace = LPyramid(lena_img, a, depth)

fig=plt.figure(dpi=180)
fig.subplots_adjust(hspace=0.4, top=0.9)
fig.suptitle('Laplacian Pyramid')

fig.add_subplot(2,4,1)
plt.imshow(Laplace[0],cmap='gray')
plt.axis("off")
plt.title("Level 0")

fig.add_subplot(2,4,2)
plt.imshow(Laplace[1],cmap='gray')
plt.axis("off")
plt.title("Level 1")

fig.add_subplot(2,4,3)
plt.imshow(Laplace[2],cmap='gray')
plt.axis("off")
plt.title("Level 2")

fig.add_subplot(2,4,4)
plt.imshow(Laplace[3],cmap='gray')
plt.axis("off")
plt.title("Level 3")
```

**5-6: Αποκωδικοποίηση και ανάκτηση αρχικής εικόνας, χωρίς και με κβαντοποίηση**

```python
def L_Pyramid_Decode(L,a):

  #Using Laplacian Pyramid of an image and parameter a
  #used for creating it, the output is the decoded image
  N = len(L) #Depth of the Pyramid
  kernel = GKernel(a) #kernel

  I_out = L[N-1]
  for n in range(N-1):
    I_out = GEXPAND(I_out, kernel)
    I_out = I_out + L[N-n-2] - 128


  return I_out

def L_Quantization(L, n):

  #Based on size of bins n, this function return the quantized value
  #for every pixel of a given level of the pyramid

  C = np.zeros(L.shape)
  
  #C list are the quantized levels

  for x in range(L.shape[0]):
    for y in range(L.shape[1]):
      m = math.ceil(L[x][y]/n - 1/2)
      C[x][y] = m*n
  
  return C

n = 5 #bin size

unq = L_Pyramid_Decode(Laplace, a)
q = []
for x in range(len(Laplace)):
  q.append(L_Quantization(Laplace[x], n))

qua = L_Pyramid_Decode(q, a)

fig=plt.figure(dpi=180)
fig.subplots_adjust(hspace=0.6, top=0.9)
fig.suptitle('Decoding')

fig.add_subplot(2,2,1)
plt.imshow(unq,cmap='gray')
plt.axis("off")
plt.title("Unquantized")

fig.add_subplot(2,2,2)
plt.imshow(qua,cmap='gray')
plt.axis("off")
plt.title("Quantized")
```
